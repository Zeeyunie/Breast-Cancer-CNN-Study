```python
indir = "./cbis-ddsm-breast-cancer-image-dataset/jpeg"

# Replace "CBIS-DDSM/jpeg" with my full indir, which is my local directory
cropped_images = cropped_images.replace("CBIS-DDSM/jpeg", indir, regex=True)
full_mammo = full_mammo.replace("CBIS-DDSM/jpeg", indir, regex=True)
roi_img = roi_img.replace("CBIS-DDSM/jpeg", indir, regex=True)
```

```python
# Map UID to image path
full_mammo_dict = {path.split("/")[3]: path for path in full_mammo}
cropped_images_dict = {path.split("/")[3]: path for path in cropped_images}
roi_img_dict = {path.split("/")[3]: path for path in roi_img}
```

```python
def fix_image_path(data):
    for index, img in enumerate(data.values):
        try:
            # Fix full mammogram path
            uid = img[11].split("/")[2]
            data.iloc[index, 11] = full_mammo_dict[uid]

            # Fix cropped image path
            uid = img[12].split("/")[2]
            data.iloc[index, 12] = cropped_images_dict[uid]

            # Fix ROI mask image path
            uid = img[13].split("/")[2]
            data.iloc[index, 13] = roi_img_dict[uid]

        except Exception as e:
            print(f"Row {index} failed to update: {e}")
```

```python
# Process all CSV files in the csv folder
import os
import glob

# Get all CSV files in the csv folder
csv_folder = "./cbis-ddsm-breast-cancer-image-dataset/csv"
csv_files = glob.glob(os.path.join(csv_folder, "*.csv"))

# Filter out dicom_info.csv and meta.csv as they don't need path fixing
case_files = [f for f in csv_files if 'case_description' in f]

print("Found case description files:")
for file in case_files:
    print(f"  - {os.path.basename(file)}")

# Process each case description file
for csv_file in case_files:
    print(f"\nProcessing {os.path.basename(csv_file)}...")
    
    # Load the CSV file
    df = pd.read_csv(csv_file)
    print(f"  Original shape: {df.shape}")
    
    # Check if the file has the required columns
    required_columns = ['image file path', 'cropped image file path', 'ROI mask file path']
    if not all(col in df.columns for col in required_columns):
        print(f"  Skipping {os.path.basename(csv_file)} - missing required columns")
        continue
    
    # Apply the path fixing function
    def fix_image_paths(data):
        failed_rows = []
        for index, img in enumerate(data.values):
            try:
                # Fix full mammogram path
                uid = img[11].split("/")[2]  # image file path is column 11
                data.iloc[index, 11] = full_mammo_dict[uid]

                # Fix cropped image path
                uid = img[12].split("/")[2]  # cropped image file path is column 12
                data.iloc[index, 12] = cropped_images_dict[uid]

                # Fix ROI mask image path
                uid = img[13].split("/")[2]  # ROI mask file path is column 13
                data.iloc[index, 13] = roi_img_dict[uid]

            except Exception as e:
                failed_rows.append((index, str(e)))
        
        return failed_rows
    
    # Fix the paths
    failed_rows = fix_image_paths(df)
    
    if failed_rows:
        print(f"  Failed to update {len(failed_rows)} rows:")
        for row_idx, error in failed_rows[:5]:  # Show first 5 errors
            print(f"    Row {row_idx}: {error}")
        if len(failed_rows) > 5:
            print(f"    ... and {len(failed_rows) - 5} more")
        
        # Drop failed rows
        failed_indices = [row[0] for row in failed_rows]
        df = df.drop(failed_indices).reset_index(drop=True)
        print(f"  Dropped {len(failed_indices)} problematic rows")
    
    # Add label column if pathology column exists
    if 'pathology' in df.columns:
        label_map = {
            'MALIGNANT': 1,
            'BENIGN': 0,
            'BENIGN_WITHOUT_CALLBACK': 0
        }
        df['label'] = df['pathology'].replace(label_map)
        print(f"  Added label column")
    
    # Save the cleaned file
    output_file = csv_file.replace('.csv', '_cleaned.csv')
    df.to_csv(output_file, index=False)
    print(f"  Saved cleaned file: {os.path.basename(output_file)}")
    print(f"  Final shape: {df.shape}")

print("\nAll files processed!")
```

```python
# Verify the results
print("Created cleaned files:")
cleaned_files = glob.glob(os.path.join(csv_folder, "*_cleaned.csv"))
for file in cleaned_files:
    print(f"  - {os.path.basename(file)}")

print(f"\nTotal cleaned files created: {len(cleaned_files)}")

# Show a summary of each cleaned file
for file in cleaned_files:
    df = pd.read_csv(file)
    print(f"\n{os.path.basename(file)}:")
    print(f"  Shape: {df.shape}")
    if 'label' in df.columns:
        print(f"  Label distribution:")
        print(f"    {df['label'].value_counts().to_dict()}")
    if 'pathology' in df.columns:
        print(f"  Pathology distribution:")
        print(f"    {df['pathology'].value_counts().to_dict()}")
```
